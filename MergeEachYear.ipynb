```python
wiki = pd.read_csv('wiki.csv')
wiki
```

```python
df_clean = pd.read_csv('2019.csv')
df_clean
```

```python
# name and 剧名 equal, merge
df_merge = pd.merge(df_clean, wiki, left_on='name', right_on='剧名', how='left')
df_merge
```

```python
# where 剧名 is not null, replace name with 剧名
df_merge[df_merge['剧名'].notnull()]
```

```python
# check duplicate
df_merge.duplicated().sum()
```

```python
# drop duplicate
df_merge.drop_duplicates(inplace=True)
```

```python
df_merge
```

```python
# drop columns
df_merge.drop(['剧名', '集数', '主演'], axis=1, inplace=True)

# change name
df_merge.rename(columns={'制作': 'shooting', '题材':'type_wiki'}, inplace=True)
```

```python
df_merge
```

```python
# saparate Premiere to year and month and day
df_merge['premiere_year'] = df_merge['Premiere'].str.split('-').str[0]
df_merge['premiere_month'] = df_merge['Premiere'].str.split('-').str[1]
df_merge['premiere_day'] = df_merge['Premiere'].str.split('-').str[2]
df_merge['首播1'] = df_merge['首播'].str.split('，').str[0]
df_merge['首播2'] = df_merge['首播'].str.split('，').str[1]

#use regex to filter 首播1 delete value
df_merge['首播1'] = df_merge['首播1'].str.split('日').str[1]
df_merge['首播2'] = df_merge['首播2'].str.split('日').str[1]

df_merge['首播1'] = df_merge['首播1'].str.split('(').str[0]
df_merge['首播2'] = df_merge['首播2'].str.split('(').str[0]
df_merge
```

```python
# change word in columns 首播1	首播2
df_merge['首播1'] = df_merge['首播1'].str.replace('优酷', '优酷视频')
df_merge['首播2'] = df_merge['首播2'].str.replace('优酷', '优酷视频')
df_merge['首播1'] = df_merge['首播1'].str.replace('Bilibili', '哔哩哔哩')
df_merge['首播2'] = df_merge['首播2'].str.replace('Bilibili', '哔哩哔哩')
df_merge
```

```python
df_merge.to_csv('merge2019.csv', index=False)
```
