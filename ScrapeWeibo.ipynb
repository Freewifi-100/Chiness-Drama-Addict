```python
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
# timestamp
import datetime
import re
import time
```

```python
df = pd.read_csv('actor_2019.csv')
df_a = pd.read_csv('actor1.csv')
# df_b = pd.read_csv('real_data/actor_weibo2.csv')
# df_c = pd.read_csv('real_data/actor_weibo3.csv')
# get only actors
actors = df['actor'].tolist()
actor_a = df_a['actor'].tolist()
# actors.extend(df['actor2'].tolist())
# actors.extend(df['actor3'].tolist())

#nerge and drop duplicates
actors.extend(actor_a)
actors = list(set(actors))


# drop name where we already have weibo account
actors = [x for x in actors if x not in actor_a]

# import real_data/actor_follower_weibo.csv
# df_new = pd.read_csv('actor_2019.csv')
# get where weibo_name is No weibo account
# df_new = df[df['weibo_name'] == 'No weibo account']
# get only actors
# actors = df_new['actor'].tolist()
# # drop duplicates
# actors = list(set(actors))

print(len(actors))
# print actors 5 `actors[:5]
print(actors[:5])

# create a new chrome session
driver = webdriver.Chrome('//chromedriver.exe')
# driver.implicitly_wait(10)
driver.get("https://m.weibo.cn")
time.sleep(2)

# find the search box
first_click = driver.find_element(By.XPATH, '//*[@id="app"]/div[1]/div[1]/div[1]/a/aside/label/div')
first_click.click()
time.sleep(2)

all_actor = []
weibo_name = []
follower = []

for i in range(len(actors)):

    # find the search box
    search_box = driver.find_element(By.XPATH, '//*[@id="app"]/div[1]/div[1]/div[1]/div/div/div[2]/form/input')
    search_box.clear()
    time.sleep(5)
    # enter search keyword and submit
    search_box.send_keys(actors[i])
    search_box.submit()
    time.sleep(5)

    try:
        # get weibo name and followers
        w = driver.find_element(By.XPATH, '//*[@id="app"]/div[1]/div[1]/div[3]/div/div/div[1]/div/div/div/div[2]/div/h3/span')
        # if this xpath null elif another xpath
        f = driver.find_element(By.CLASS_NAME, 'm-text-box')
        # f = re.findall(r'粉丝：(\w+万)|(\w+.)', f.text)
        # if f == '':
        #     f = driver.find_element(By.XPATH, '//*[@id="app"]/div[1]/div[1]/div[3]/div/div/div[1]/div/div/div/div[2]/div/h4[2]')
        
    
        print(i, actors[i])
        print(w.text, f.text)

        all_actor.append(actors[i])
        weibo_name.append(w.text)
        follower.append(f.text)
    
    except NoSuchElementException:
        print('No weibo account')
        all_actor.append(actors[i])
        weibo_name.append('No weibo account')
        follower.append('No weibo account')
    
    # random refresh
    time.sleep(5)

# Close the browser
driver.quit()
```

```python
# create a new dataframe
df_new = pd.DataFrame({'actor': all_actor, 'weibo_name': weibo_name, 'follower': follower, 'weibo_score': follower})
```
