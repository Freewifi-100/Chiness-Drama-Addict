```python
import pandas as pd
import numpy as np
```

```python
# marge 2020 -2022
df_2019 = pd.read_csv('data/2019.csv')
df_2020 = pd.read_csv('data/2020.csv')
df_2021 = pd.read_csv('data/2021.csv')
df_2022 = pd.read_csv('data/2022.csv')
df_2023 = pd.read_csv('data/2023.csv')

# concat
df = pd.concat([df_2019, df_2020, df_2021, df_2022, df_2023], ignore_index=True)
df.head()
```

```python
# check duplicates
df.duplicated().sum()
```

```python
#save
df.to_csv('data/original_data.csv', index=False)
```

```python
# rop many col
def drop_col(df, col_list):
    df = df.drop(columns=col_list)
    return df

def order_col(df, col_list):
    df = df[col_list]
    return df
```

```python
# # # split writer column into 3 columns and have nan if no writer
import re
def clean(df):

    df['name'] = df['name'].astype(str)
    df['name_s'] = df['name'].apply(lambda x: x.split(' ', 1)[0] if isinstance(x, str) else x)
    df['name_s'] = df['name_s'].apply(lambda x: x.split('·', 1)[0] if isinstance(x, str) else x)
    df['director'] = df['director'].str.split(' / ').str[0]

    df['writer'] = df['writer'].str.split(' / ').str[0]

    # split actor column into 4 columns and have nan if no actor
    df['actor1'] = df['actor'].str.split(' / ').str[0]
    df['actor2'] = df['actor'].str.split(' / ').str[1]
    df['actor3'] = df['actor'].str.split(' / ').str[2]

    # split genre column into 5 columns and have nan if no genre
    df['genre1'] = df['genre'].str.split(' / ').str[0]
    df['genre2'] = df['genre'].str.split(' / ').str[1]
    df['genre3'] = df['genre'].str.split(' / ').str[2]
    df['genre4'] = df['genre'].str.split(' / ').str[3]
    df['genre5'] = df['genre'].str.split(' / ').str[4]
    
    # split aka column into 3 columns and have nan if no aka
    df['aka1'] = df['aka'].str.split(' / ').str[0]
    df['aka2'] = df['aka'].str.split(' / ').str[1]
    df['aka3'] = df['aka'].str.split(' / ').str[2]

    df['country'] = df['country'].str.split(' / ').str[0]


    df['language'] = df['language'].str.split(' / ').str[0]

    # chose =word after :
    df['season'] = df['season'].str.split(': ').str[1]
    df['director'] = df['director'].str.split(': ').str[1]
    df['country'] = df['country'].str.split(': ').str[1]
    df['language'] = df['language'].str.split(': ').str[1]
    df['Premiere'] = df['Premiere'].str.split('(').str[0]
    df['Episode'] = df['Episode'].str.split(': ').str[1]
    df['episode_duration'] = df['episode_duration'].str.split(': ').str[1]
    df['IMDb'] = df['IMDb'].str.split(': ').str[1]
    df['aka1'] = df['aka1'].str.split(':').str[1]
    df['writer'] = df['writer'].str.split(':').str[1]
    df['genre1'] = df['genre1'].str.split(':').str[1]
    #get review only number
    df['review'] = df['review'].astype(str).str.findall(r'\d+').str[0]
    
```

```python
# df = drop_col(df, ['web-scraper-order', 'web-scraper-start-url'])
df = df.rename(columns={'link-href': 'uniq_id', 'Episode Length': 'episode_duration'})
# df = order_col(df, ['uniq_id', 'name', 'season', 'director', 'writer', 'actor', 'genre', 'country', 'language', 'Premiere', 'Episode', 'episode_duration', 'aka', 'IMDb', 'score', 'review', '5star', '4star', '3star', '2star', '1star'])
clean(df)
```
