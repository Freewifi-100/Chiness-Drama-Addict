```python
import pandas as pd
import numpy as np
```

```python
check  = pd.read_csv('check.csv')
df_clean = pd.read_csv('ChineseDrama_clean.csv')


# merge the two dataframes
df = pd.merge(df_clean, check, left_on='uniq_id', right_on='web-scraper-start-url', how='left')
df
```

```python
#colaesce channel and channel_x to channel and permission and permission_x to permission
df['channel'] = df['channel_x'].combine_first(df['channel_y'])
df['permission'] = df['permission_x'].combine_first(df['permission_y'])
df.drop(['channel_x', 'permission_x', 'channel_y', 'permission_y','name_y','web-scraper-start-url','web-scraper-order'], axis=1, inplace=True)
df
```

```python
df.to_csv('ChineseDrama_clean.csv', index=False)
```
